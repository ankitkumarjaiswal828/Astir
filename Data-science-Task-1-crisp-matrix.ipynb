```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_datasets  = pd.read_csv('training.csv')
train_datasets.head(5)
```

```python
train_datasets.info()
```

```python
train_datasets.describe()
```

```python
plt.figure(figsize = (30,10))
sns.heatmap(train_datasets.corr(),annot = True)
```

```python
test_datasets = pd.read_csv('testing.csv')
test_datasets.head(5)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(test_datasets.corr(),annot = True)
```

```python
x_train= train_datasets.drop(columns=['prognosis']).dropna(axis =1,how ='any',thresh =None,subset = None,inplace = False)
y_train =train_datasets['prognosis']
```

```python
x_test = test_datasets.drop(columns = ['prognosis'])
y_test = test_datasets['prognosis']
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
tree = DTC()
tree.fit(x_train,y_train)
```

```python
tree_Prediction = tree.predict(x_test)
tree_Prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
acuraccy_tree = accuracy_score(tree_Prediction,y_test)
print('Dtree_accruracy  ' + format(acuraccy_tree))
```

```python
lr = LR()
lr.fit(x_train,y_train)
```

```python
lr_prediction = lr.predict(x_test)
lr_prediction
```

```python
accuracy_Log = accuracy_score(lr_prediction,y_test)
print('logistic_regression_accuracy:  '+ format(accuracy_Log))
```

```python
rfc = RFC()
rfc.fit(x_train,y_train)
```

```python
rfc_prediction = rfc.predict(x_test)
rfc_prediction
```

```python
accuracy_rfc = accuracy_score(rfc_prediction,y_test)
print('random_forest_accuracy ' + format(accuracy_rfc))
```

```python
acc_list=[acuraccy_tree,accuracy_Log,accuracy_rfc]
import matplotlib.pyplot as plt
plt.rcdefaults() 
objects = ('Decision Tree','Logistic Regression', 'Random Forest')
y_pos = np.arange(len(objects))
performance = acc_list
width = 1/5
plt.barh(y_pos, performance, width,align='center', alpha=1)
plt.yticks(y_pos,objects)
plt.xlabel('Accuracy')
plt.xlabel('Algorithms')
plt.title('Accuracy Scores of Algorithms')
plt.show()
```

Thats way Decision tree and Random forest classifier used on disease training datasets so given 97.6% accuracy ,
